```python
#Basics of Pairwise
from Bio import Align
from Bio import pairwise2                                    #for match, mismatches, gap-based score use pairwise2
aligner = Align.PairwiseAligner()
aligner.match_score= 1.0

# aligner = Align.Pairwiser(match_score= 1.0), this method was given by sir check this if the above doesn't work.
```

```python
# Defining the sequence for alignment
SeqA = "GATCTTAGCC"
SeqB = "GCTCTTACCGAAUG"

#Call align to score
score = aligner.score(SeqA, SeqB)
score
```

```python
# Aligner.align methods return Alinment obj, each rep 1 alignmnet between the 2 sequences
alignments = aligner.align(SeqA, SeqB)
alignment = alignments[0]
alignment  
```

```python
#To visualize the alignments
for alignment in alignments:
    print(alignment)
```
