```python
#1. Define your protein sequence
sequence_id = "protein1"
description = "Human Hb Subunit"
sequence = "MVHLTPEEKSAVTALWGKVNVDEVGGEALGRLLVVYPWTQRFFESFGDLSTPDAVMGNPKVKAHGKKVLGAFSDGLAHLDNLKGTFATLSELHCDKLHVDPENFRLLGNVLVCVLAHHFGKEFTPPVQAAYQKVVAGVANALAHKYH"
```

```python
#2. Define the output FASTA file name (HHS- Human Haemoglobin Subunit)
filename= "protein_sequenceHHS.fasta" 
```

```python
#3. Write to the file 
with open(filename, "w") as fasta_file:
    #Header line starting with ">"
    fasta_file.write(f">{sequence_id} {description}\n")  # Fixed: |n to \n for newline
    # Sequence lines (optional: wrap at 60 chr per line)
    for i in range(0, len(sequence), 60):  # Fixed: Indented this loop inside the with block
        fasta_file.write(sequence[i: i+60] + "\n")
        
print(f"Protein Sequence saved to {filename}")
```
