```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# file path
file_path = "C:/Users/steff/OneDrive/Desktop/Rough R-scripts/GSE183947_fpkm.csv"
```

```python
#Read the csv file
df = pd.read_csv(file_path)
```

```python
# Show the first 5 rows
print(df.head())
```

```python
# Get Basic Info
print(df.info())

# Summary statistics
print(df.describe())

# Check for missing values 
print(df.isnull().sum())

# List column names
print(df.columns.tolist())
```

```python
# Set gene names as index
df.set_index(df.columns[0], inplace = True)

#Check new shape
print("Shape:", df.shape)

#Covert all daata to numeric (in case there are any non-numeric columns)
df = df.apply(pd.to_numeric, errors = 'coerce')
```

```python
# Basic filtering: remove genes with very low average expression
df_filtered = df[df.mean(axis=1) > 1]
print("Filterd Shape:", df_filtered.shape)
```

```python
# Flatten all values into a single list
all_values = df_filtered.values.flatten()

plt.figure(figsize=(8, 5))
sns.histplot(all_values, bins= 100, kde= True)
plt.title("Distribution of FPKM Values")
plt.xlabel("FPKM")
plt.ylabel("Count")
plt.show()
```

```python
# Compute variance for each gene
df_filtered["var"] = df_filtered.var(axis= 1)

# Select top 50 variables genes
top_var_genes = df_filtered.sort_values("var", ascending = False).head(50)
top_var_genes.drop(columns= ["var"], inplace= True)

# Plot heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(top_var_genes, cmap= "viridis")
plt.title("Top 50 Variable Genes")
plt.xlabel("Samples")
plt.ylabel("Genes")
plt.show()
```
