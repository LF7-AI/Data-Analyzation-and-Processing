```python
# Writing a FASTA file
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO

seq_obj = SeqRecord(Seq("ATGCATGACTGA"), id = "Test",
                    description = "Example sequence")
SeqIO.write(seq_obj, "output.fasta", "fasta")
```

```python
# To calculate the GC count
from Bio import SeqIO

def calculate_gc(seq):
    g = seq.count("G")
    c = seq.count("C")
    gc_content = 100*float(g+c)/len(seq)
    return gc_content

for record in SeqIO.parse("gene.fna", "fasta"):
    print(f"{record.id}: {calculate_gc(record.seq):.2f}%GC")
    
```
