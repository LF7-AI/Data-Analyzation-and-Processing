```python
# Import Required Libraries
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# Step 1: Generate synthetic EHR-like dataset
np.random.seed(42)
n_samples = 200
ehr_data = pd.DataFrame({
    'age': np.random.randint(30, 80, size = n_samples),
    'gender': np.random.choice(['Male', 'Female'], size = n_samples),
    'blood_pressure': np.random.normal(120, 15, size= n_samples),
    'cholesterol': np.random.normal(200, 30, size= n_samples),
    'glucose': np.random.normal(100, 20, size= n_samples),
    'diagnosis': np.random.choice([0, 1], size= n_samples)
})

# Watch the generated dataset
print(ehr_data.head())
print(ehr_data.info())
print(ehr_data.describe())
```

```python
# Step 2: Introduce missing values
mask = np.random.rand(n_samples) < 0.2
ehr_data.loc[mask, 'cholesterol'] = np.nan
mask = np.random.rand(n_samples) < 0.15
ehr_data.loc[mask, 'glucose'] = np.nan

# Watch the dataset after introducing missing values
print("\nDataset after introducing missing values:")
print(ehr_data.head())
print(ehr_data.info())
print(ehr_data.isnull().sum())
```

```python
#Step 3: Impute missing values using KNNImputer
imputer = KNNImputer(n_neighbors= 5)
ehr_data[['cholesterol', 'glucose']] = imputer.fit_transform(ehr_data[['cholesterol', 'glucose']])

# Watch the dataset after imputation
print("\nDataset after imputing missing values:")
print(ehr_data.head())
print(ehr_data.isnull().sum())
```

```python
# Step 4: Encode categorical features
label_encoder = LabelEncoder()
ehr_data['gender'] = label_encoder.fit_transform(ehr_data['gender']) # Male:1, Feamle:0

# Watch the dataset after encoding
print("\nDataset after encoding categorical features:")
print(ehr_data.head())
print(ehr_data.info())
```

```python
# Step 5: Impute missing values using KNNImputer (actually step 3, but here we are using columns). 
#This current code imputes all numerical columns in ehr_data that contain missing values, 
#creating a completely new DataFrame ehr_imputed with no NaNs (where possible).

imputer = KNNImputer(n_neighbors= 5)
ehr_imputed = pd.DataFrame(imputer.fit_transform(ehr_data), 
                           columns = ehr_data.columns)

# Step 6: Correlatiion heatmap after imputation
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (8, 6))
sns.heatmap(ehr_imputed.corr(), annot= True, cmap= 'coolwarm', fmt= '.2f')
plt.title('Feature Correlation Matrix')
plt.show()
```

```python
# Step 6: Feature-target split
X = ehr_imputed.drop('diagnosis', axis =1)
Y = ehr_imputed['diagnosis']

# Step 7: Train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, 
                                                    test_size = 0.2, random_state=42)

#Prompt to view the dataset again
print("\nOringinal generated dataset:")
print(ehr_data.head())
print(ehr_data.info())
print(ehr_data.describe())
```

```python
# Step 8: Train a classifier (e.g., RandomForestClassifier)
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, Y_train)
RandomForestClassifier(random_state=42)

# Step 9: Evaluate the model
y_pred = clf.predict(X_test)
print(classification_report(Y_test, y_pred))
```

```python
# Step 11: Fature importance visualization
importances = clf.feature_importances_
features = X.columns
indices = np.argsort(importances)[::-1]

plt.figure(figsize= (8, 6))
sns.barplot(x = importances[indices], y = features[indices], palette = "magma")
plt.title('Feature Importances from Random Forest')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.tight_layout()
plt.show()
sns.barplot(x = importances[indices], y= features[indices], palette= "magma")
```

```python
# Step 12: ROC Curve
from sklearn.metrics import roc_curve, auc
probs = clf.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(Y_test, probs)
roc_auc = auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, color= 'darkorange', lw= 2, label= 'ROC Curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color= 'navy', lw = 2, linestyle= '--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc= "lower right")
plt.show()
```
